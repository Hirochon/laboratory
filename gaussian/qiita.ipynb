```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
```

```python
#関数に投入するデータを作成
x = y = np.arange(-20, 20, 0.5) # 0.5ごとに-20~20までを出力 
X, Y = np.meshgrid(x, y) # 格子点をきめて増やす

z = np.c_[X.ravel(),Y.ravel()] # X.ravel()で１次元に変換して、np.c_[]で２つの引数の２次元へ変換される。
```

```python
#二次元正規分布の確率密度を返す関数
def gaussian(x, sigma, mu):
    #分散共分散行列の行列式
    det = np.linalg.det(sigma) # 行列式を求める
    print(det)
    #分散共分散行列の逆行列
    inv = np.linalg.inv(sigma) # 逆行列を求めてる
    n = x.ndim # xの次元
    print(inv)
    return np.exp(-np.diag((x - mu)@inv@(x - mu).T)/2.0) / (np.sqrt((2 * np.pi) ** n * det)) # 
```

```python
#2変数の平均値を指定
mu = np.array([0,0])
#2変数の分散共分散行列を指定
sigma = np.array([[50,0],[0,50]])

Z = gaussian(z, sigma, mu)
shape = X.shape
Z = Z.reshape(shape)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(Z)
plt.show()
```

```python
#二次元正規分布をplot
fig = plt.figure(figsize = (15, 15))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm)
plt.show()
```
