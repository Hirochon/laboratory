```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian(x, y, a, b, zure_x, zure_y, theta):
    
    rotate = np.array([[np.cos(theta),np.sin(theta)],[-np.sin(theta), np.cos(theta)]])
    
    x2 = (x - zure_x)
    y2 = (y - zure_y)
    
    [X, Y] = np.dot([x2, y2], rotate)
    
    XX = X**2
    YY = Y**2
    aa = a**2
    bb = b**2
    
    return np.exp(-(XX/aa) - (YY/bb))
```

```python
data = []
for k in range(16):
    init_ar = np.zeros(2500).reshape(50,50)
    theta = np.random.rand() * 2 * np.pi
    a = np.random.randint(1, 30)
    b = np.random.randint(1, 30)
    x_zure = np.random.randint(-25, 25)
    y_zure = np.random.randint(-25, 25)
    
    for i in range(50):
        for j in range(50):
            y = i - 25
            x = j - 25
            init_ar[i, j] = gaussian(x, y, a, b, x_zure, y_zure, theta)
    data.append(init_ar)
```

```python
fig = plt.figure(figsize=(15,15))
for k in range(16):
    fig.add_subplot(4, 4, k+1)
    plt.imshow(data[k])
plt.show()
```
