```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4, 4, 2000)
```

```python
plt.figure()
plt.gca().yaxis.set_tick_params(which='both', direction='in')
plt.gca().xaxis.set_tick_params(which='both', direction='in')
plt.minorticks_on()
plt.plot(x, np.tanh(x))
plt.show()
```

```python
plt.figure()
plt.gca().yaxis.set_tick_params(which='both', direction='in')
plt.gca().xaxis.set_tick_params(which='both', direction='in')
plt.minorticks_on()
plt.plot(x, np.maximum(0, x))
plt.show()
```

```python
plt.figure()
plt.gca().yaxis.set_tick_params(which='both', direction='in')
plt.gca().xaxis.set_tick_params(which='both', direction='in')
plt.minorticks_on()
plt.plot(x, np.where(x > 0, x , 0.3 * x))
plt.show()
```

```python
plt.figure()
plt.gca().yaxis.set_tick_params(which='both', direction='in')
plt.gca().xaxis.set_tick_params(which='both', direction='in')
plt.minorticks_on()
plt.plot(x, np.where(x > 0, x , 0.03 * x))
plt.show()
```
